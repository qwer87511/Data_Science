### Show the simple statistical information (mean, std, min, max, quartile1, quartile2, quartile3).

```python
#def percentile(n):
#    def percentile_(x):
#        return x.quantile(n)
#    percentile_.__name__ = 'percentile_{:2.0f}'.format(n*100)
#    return percentile_
def quartile(n):
    def percentile_(x):
        return x.quantile(0.25*n)
    percentile_.__name__ = 'quartile{:d}'.format(n)
    return percentile_
df.agg([np.mean, np.std, np.min, np.max, quartile(1), quartile(2), quartile(3)])
```

# Step 6 Use matplotlib show 2D images about data.
[Matplotlib](https://matplotlib.org/)
### Plot the distribution of two class (1. GDP < 58000, 2. GDP >= 58000) of COVID-19 State Data using different colors and different  marker where x-axis is the Pollution and y-axis the Mortality-rate.

```python
%matplotlib inline
import matplotlib.pyplot as plt

df['Mortality-rate']=[d / i for d, i in zip(df['Deaths'], df['Infected'])]

plt.xlabel('Pollution')
plt.ylabel('Mortality-rate')
plt.scatter(df['Pollution'], df['Mortality-rate'], c = [0 if gdp < 58000 else 1 for gdp in df['GDP']])
plt.show()
```
