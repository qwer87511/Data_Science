## User Input and Function

```python
#Exercise: Write a function to find prime numbers between a given range [n, m]
n = int(input("n: "))
m = int(input("m: "))
answers = print_prime_number(n,m)
for answer in answers:
    print(answer)
```

```python
#Answer:

#We can use "break" and "continue" to improve the efficiency of the program
#The following code finds the prime numbers within [n,m]

def print_prime_number(n,m):
    answers=[]
    for i in range(n,m+1):  #set the loop range from n to m, including both n and m
        prime=True          #assume each number is prime number without testing
        for j in range(2,i-1):   #create an inner loop to find if there is a factor (quiz: what's a better range than this)
            if i%j ==0:          #if a factor is found
                prime=False      #mark i is not prime
                break            #completely stop inner loop
        if prime==False:    #we can simple use if prime: print(i) here. But I'd like to show the use continue
            continue        #skip the remaining statements of i loop

        answers.append(i)            # add the new found to answer list 
        
    return answers
```

## Basic IO

```python
#write answer to a textfile 
with open('data.txt', 'w') as f:
    for answer in answers:
        f.write("%s\n" % answer)
```

```python
# I will demonstrate a simple data science project using the folloing 
# python code. You don't have to understand all the functions and packages at this point of time

# import the needed packages
import sklearn
from sklearn import datasets
from sklearn import linear_model
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#load the data
boston = datasets.load_boston()
```

```python
print(boston.DESCR)
```

```python
boston.feature_names
```

```python
DF_boston = pd.DataFrame(boston.data)
DF_boston.head()
```

```python
DF_boston.columns=boston.feature_names
DF_boston.head()
```

```python
DF_boston['Price'] = boston.target
DF_boston.head()
```

```python
sns.lmplot('RM','Price',data=DF_boston)
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(DF_boston['RM'].values.reshape(-1,1),DF_boston['Price'])
```

```python
print('Intercept: \n', model.intercept_)
print('Coefficients: \n', model.coef_)
```
