# Step 1
### Using Pandas load Iris Data Set as dataframe from UCI Machine Learning Repository.  
**[Pandas](https://pandas.pydata.org/)**  
**[Iris Data Set](https://archive.ics.uci.edu/ml/datasets/iris)**

```python
import pandas as pd 

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',
                 header=None,
                 names=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'class'])
```

# Step 7
### Using matplotlib show the 2D image of data.
[Matplotlib](https://matplotlib.org/)

```python
%matplotlib inline
import matplotlib.pyplot as plt

colorDict = {
    'Iris-virginica': 0,
    'Iris-versicolor': 1,
    'Iris-setosa': 2
}

plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c = [colorDict[c] for c in df['class']])
```

### Show the distrubution of three type Iris data.

```python
plt.pie(
    [len(df[df['class'] == 'Iris-virginica']), len(df[df['class'] == 'Iris-versicolor']), len(df[df['class'] == 'Iris-setosa'])],
    explode=[0.05, 0, 0],
    labels = ['Iris-virginica', 'Iris-versicolor', 'Iris-setosa'],
    autopct= "%3.1f%%")
```

# Step 8
### Using PCA reduce dimension of data from 4 to 2 by scikit-learn and show 3D imgage.
[scikit-learn](https://scikit-learn.org/stable/)

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 

reduced = PCA(n_components=3).fit_transform(df.iloc[:, 0:4])

fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)

ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.set_ylabel("2nd eigenvector")
ax.set_zlabel("3rd eigenvector")

ax.scatter(reduced[:, 0], reduced[:, 1], reduced[:, 2], c = [colorDict[c] for c in df['class']])
```
